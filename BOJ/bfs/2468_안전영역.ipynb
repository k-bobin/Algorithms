```python
import sys
from collections import deque

input = sys.stdin.readline
sys.setrecursionlimit(100000)

dx = [-1, 0, 1, 0]
dy = [0, -1, 0, 1]

def bfs(y, x, h):
    queue = deque([(y, x)])

    while queue:
        cy, cx = queue.popleft()

        for k in range(4):
            ny = cy + dy[k]
            nx = cx + dx[k]

            if 0 <= nx < N and 0 <= ny < N and map_data[ny][nx] > h and not chk[ny][nx]:
                chk[ny][nx] = True
                queue.append((ny, nx))

N = int(input())
map_data = [list(map(int, input().strip().split())) for _ in range(N)]
num = 1

for p in range(1, 101):
    chk = [[False] * N for _ in range(N)]
    cnt = 0
    for j in range(N):
        for i in range(N):
            if map_data[j][i] > p and not chk[j][i]:
                chk[j][i] = True
                bfs(j, i, p)
                cnt += 1
    num = max(num, cnt)

print(num)

```
